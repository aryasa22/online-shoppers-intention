# 2. Univariate Analysis

Gunakan visualisasi untuk melihat distribusi masing-masing kolom (feature maupun target).<br> 
Tuliskan hasil observasinya, misalnya jika ada suatu kolom yang distribusinya menarik (misal skewed, bimodal, ada outlier, ada nilai yang mendominasi, kategorinya terlalu banyak, dsb). Jelaskan juga apa yang harus di-follow up saat data pre-processing.

```python
## Numerical analysist
```

```python
features = numericals
for i in range(0,len(features)):
    plt.subplot(2,5,i+1)
    sns.boxplot(y=df[features[i]],color='blue',orient='v')
    plt.tight_layout()
```

**Berapa banyak dan berapa jauh outlier di setiap kolom**

```python
for i in range(0,len(features)):
    plt.subplot(5,2,i+1)
    sns.kdeplot(x=df[features[i]],color='green')
    plt.xlabel(features[i])
    plt.tight_layout()
```

**bagaimana bentuk distribusi setiap kolom.(normal,skewed,bimodal)<br>
apakah ada nilai tertentu yang umum.**

```python
for i in range(0,len(features)):
    plt.subplot(2,5,i+1)
    sns.violinplot(y=df[features[i]],color='blue',orient='v')
    plt.tight_layout()
```

```python
## categoricals analysis
```

```python
for i in range(0,len(categoricals)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[categoricals[i]],color='green')
    plt.tight_layout()
```

**ketimpangan antar kategori, terutama pada target<br>
sebaran kategori yang timpang pada feature merupakan indikasi ketidakgunaan feature. pada targer, sebaran yang timpang dapat membuat proses learning gagal.**

```python
for i in range(0,len(encoded_cat)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[encoded_cat[i]],color='green')
    plt.tight_layout()
```
